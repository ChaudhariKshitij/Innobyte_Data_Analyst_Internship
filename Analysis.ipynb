```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("SampleSuperstore.csv")
```

```python
df
```

## Data Cleaning

```python
df=df.dropna()
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
profit=df[df['Profit']>0]
profit
```

```python
profit[['Sales','Quantity','Discount','Profit']].describe()
```

```python
profit.groupby(by='City').sum().sort_values('Profit',ascending=True).head(10)
```

```python
loss=df[df['Profit']<0]
loss
```

```python
loss[['Sales','Quantity','Discount','Profit']].describe()
```

```python
total_sales = df['Sales'].sum()
total_sales
```

```python
sales_by_category = df.groupby('Category')['Sales'].sum()
sales_by_category
```

```python
sales_by_subcategory = df.groupby('Sub-Category')['Sales'].sum()
sales_by_subcategory
```

```python
total_profit = df['Profit'].sum()
total_profit
```

```python
profit_by_category = df.groupby('Category')['Profit'].sum()
profit_by_category
```

```python
profit_by_subcategory = df.groupby('Sub-Category')['Profit'].sum()
profit_by_subcategory
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum()
segment_sales
```

```python
segment_profit = df.groupby('Segment')['Profit'].sum()
segment_profit
```

```python
sales_by_country = df.groupby('Country')['Sales'].sum()
sales_by_country
```

```python
profit_by_region = df.groupby('Region')['Profit'].sum()
profit_by_region
```

```python
top_categories = df.groupby('Category')['Sales'].sum().nlargest(5)
top_categories
```

```python
profitable_categories = df.groupby('Category')['Profit'].sum().nlargest(5)
profitable_categories
```

```python
discount_impact = df.groupby('Discount')[['Sales', 'Profit']].mean()
discount_impact
```

```python
ship_mode_sales = df.groupby('Ship Mode')['Sales'].sum()
ship_mode_sales
```

```python
ship_mode_profit = df.groupby('Ship Mode')['Profit'].sum()
ship_mode_profit
```

```python
quantity_sold = df['Quantity'].sum()
quantity_sold
```

```python
print("Total Sales:", total_sales,'\n')
print("Sales by Category:\n", sales_by_category,'\n')
print("Sales by Sub-Category:\n", sales_by_subcategory,'\n')
print("Total Profit:", total_profit,'\n')
print("Profit by Category:\n", profit_by_category,'\n')
print("Profit by Sub-Category:\n", profit_by_subcategory,'\n')
print("Segment-wise Sales:\n", segment_sales,'\n')
print("Segment-wise Profit:\n", segment_profit,'\n')
print("Sales by Country:\n", sales_by_country,'\n')
print("Profit by Region:\n", profit_by_region,'\n')
print("Top Categories by Sales:\n", top_categories,'\n')
print("Top Categories by Profit:\n", profitable_categories,'\n')
print("Discount Impact:\n", discount_impact,'\n')
print("Ship Mode-wise Sales:\n", ship_mode_sales,'\n')
print("Ship Mode-wise Profit:\n", ship_mode_profit,'\n')
print("Total Quantity Sold:", quantity_sold)
```

```python
plt.figure(figsize=(10, 6))
sales_by_category.plot(kind='bar', color='skyblue')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
profit_by_category.plot(kind='bar', color='lightgreen')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(segment_sales, labels=segment_sales.index, autopct='%1.1f%%', startangle=140)
plt.title('Segment-wise Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
discount_impact.plot(kind='line')
plt.title('Discount Impact')
plt.xlabel('Discount')
plt.ylabel('Average Sales/Profit')
plt.xticks(discount_impact.index)
plt.grid(True)
plt.show()
```

```python
numerical_features=df[['Sales','Quantity','Discount','Profit']]
correlation_matrix=numerical_features.corr()

plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title('Correlation Matrix')
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
plt.show()
```
